```python
#Normalising the data
import re 
import numpy as np
def find_no(text):
    n = re.findall(r'[0-9]+',text)
    return "".join(n)

data = pd.read_csv("Cars.csv")

data['Make/Model'] = data['Make/Model'].str.replace("Alfa Romeo","Alfa-Romeo")
data['Make/Model'] = data['Make/Model'].str.replace("Land Rover","Land-Rover")
data['Make/Model'] = data['Make/Model'].str.replace("Aston Martin","Aston-Martin")
data['Make/Model'] = data['Make/Model'].str.replace("Mercedes-Benz","Mercedes")

new = data['Make/Model'].str.split(" ", expand = True)
data['Make'] = new[0]
data['Model'] = new[1]

#data.head()
data.drop(['Make/Model'],axis = 1)

data['Price'] = data['Price'].str.replace('€','')
data['Price'] = pd.to_numeric(data['Price'])
data['Price'].to_csv("data1.csv", header=["Price"], index = False)

data['Transmission'] = data['Transmission'].apply(lambda x: find_no(x))
data['Transmission'].replace('',0,inplace=True)

data['Fuel'] = data['Fuel'].fillna('Fuel/Gasoline/diesel/Automatic')

data['Power'] = data['Power'].fillna(0)

data['Engine CC']= data['Engine CC'].str.replace('cc','')
data['Engine CC']= data['Engine CC'].str.replace('n.b.','')
data['Engine CC']= data['Engine CC'].str.replace('nm','')
data['Engine CC'].replace('',0,inplace=True)

data['Price'] = data['Price'].astype(float)
data['Transmission'] = data['Transmission'].astype(float)
data['Engine CC'] = data['Engine CC'].astype(float)
data['Power'] = data['Power'].astype(float)

data.to_csv("car1.csv")
```

```python

makes = list(data.Make.unique())
unique = list(data.Model.unique())
all_data = list(data['Model']) 

make = []
model =[]
price = []
transmission = []
engine = []
power = []
fuel_type = []

for i in range(len(unique)):
    prices = []
    transmissions = []
    engines = []
    powers = []
    fuels = []
    models = []
    makes = []
    for j in range(len(all_data)):
        if (unique[i]) == (all_data[j]):
            
            makes.append(data['Make'][j])  
            prices.append(data['Price'][j])
            transmissions.append(data['Transmission'][j])
            engines.append(data['Engine CC'][j])
            powers.append(data['Power'][j])
            fuels.append(data['Fuel'][j])

    make.append(makes[0])        
    model.append(unique[i])
    price.append(sum(prices)/len(prices))
    transmission.append(sum(transmissions)/len(transmissions))
    engine.append(sum(engines)/len(engines))
    power.append(sum(powers)/len(powers))
    fuel_type.append(max(fuels,key=fuels.count))
```
